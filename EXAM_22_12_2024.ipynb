```python
data = [
    {"cust_id": 1, "timestamp": "2024-01-01", "residence": "Tel Aviv", "grade": 8},
    {"cust_id": 2, "timestamp": "2024-01-02", "residence": "Jerusalem", "grade": None},
    {"cust_id": 1, "timestamp": "2024-01-03", "residence": "Tel Aviv", "grade": 8},
    {"cust_id": 3, "timestamp": "2024-01-04", "residence": None, "grade": 7},
    {"cust_id": 2, "timestamp": "2024-01-05", "residence": "Jerusalem", "grade": None}
]

df = pd.DataFrame(data);
df
```

```python
# כתוב קוד )ללא שימוש בספריית PD )אשר מחשב את ממוצע הציוניםvומחליף את ציוני ה- None בממוצע  
```

```python
df = pd.DataFrame(data);
df_unique = df.drop_duplicates(subset=["cust_id"], keep="first");
df_no_none = df_unique.dropna();

average_grade = df_no_none["grade"].mean();
df_no_none["grade"] = df_no_none["grade"].fillna(average_grade);

print("Data after tasks in Pandas:");
df_no_none
```

```python
# השתמש בפקודת apply ו - groupby והחלף את ציוני ה- None בממוצע של אותו העיר )לדוגמא ברשומה שחסר בה ציון והתושב גר בתל-אביב, יוחלף הציון None בממוצע של תל אביב
```

```python
data = pd.DataFrame(data);
city_mean = data.groupby("residence")["grade"].mean();
display(city_mean);
data["grade"] = data.apply(lambda row: city_mean[row["residence"]] if pd.isna(row["grade"]) else row["grade"], axis=1);
display(data);
```
